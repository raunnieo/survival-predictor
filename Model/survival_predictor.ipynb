#**Predicting passenger survival on the Titanic**

###Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

###Loading the Dataset

```python
CL_data= pd.read_csv("dataset.csv")
CL_data.head(5)
#CL_data.shape
#CL_data.describe()
```

###Selecting the features for regression

```python
X = CL_data.iloc[:, 0: 13].values
y = CL_data.iloc[:, 13].values
```

###Checking for any NULL values in features

```python
pd.DataFrame(X).isnull().sum()
pd.DataFrame(y).isnull().sum()
```

###Handling Null Values

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
list2=[X_train.shape,X_test.shape,y_train.shape, y_test.shape]
list2
#pd.DataFrame(X_train).head(5)
```

###Imputation of missing values


```python
from sklearn.impute import SimpleImputer
import numpy as np

# Create an instance of SimpleImputer with strategy='mean'
imputer = SimpleImputer(strategy='mean')

# Fit and transform the imputer on X_train
X_train = imputer.fit_transform(X_train)

# Transform X_test using the fitted imputer
X_test = imputer.transform(X_test)

# Now, X_train and X_test should not have any missing values and are imputed with the mean
```

###Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train
```

###Training the model


```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

###Predicting the Survival

```python
y_pred = classifier.predict(X_test)
```

###Evaluating Model

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python

tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
print("True Negative=",tn," False Positive=", fp," False Negative=", fn,"True Positive=", tp)
```

```python
from sklearn.metrics import accuracy_score
acc = (accuracy_score(y_test, y_pred)*100)
print ("Accuracy = ", acc)
```
